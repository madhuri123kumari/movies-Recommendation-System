```python
import ast

def convert(obj):
    L = []
    # ast.literal_eval converts the string '[{"name": "Action"}]' into a real Python list
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert(obj):
    L = []
    counter = 0
    for i in ast.literal_eval(obj):
        if counter != 3:
            L.append(i['name'])
            counter += 1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert)
```

```python
movies.head()
```

```python
movies['crew'][0]
```

```python
def fetch_director(obj):
    L = []
    # ast.literal_eval converts the string '[{"name": "Action"}]' into a real Python list
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])  # <--- Indented inside the if
            break                # <--- Indented inside the if
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies['genres'] = movies['genres'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['cast'] = movies['cast'].apply(lambda x: [i.replace(" ", "") for i in x])
movies['crew'] = movies['crew'].apply(lambda x: [i.replace(" ", "") for i in x])
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
def stem(text):
    y = []
    for i in text.split():
        y.append(ps.stem(i))  # <--- Changed 1 to i
    return " ".join(y)        # <--- Aligned perfectly with 'for'
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```
